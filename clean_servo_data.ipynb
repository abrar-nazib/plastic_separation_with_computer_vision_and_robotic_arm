```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Open the file
df = pd.read_csv('servo_data_not_clean.csv')
df = df.dropna()
df.head(79)
```

```python
# Get the index of the row that has 76:145:4 as Angles
index = df[df['Angles'] == '86:153:0'].index[0]
index
```

```python
# Create 5 new columns: S1, S2, S3, XX, YY
df['S1'] = df['Angles'].apply(lambda x: x.split(':')[0])
df['S2'] = df['Angles'].apply(lambda x: x.split(':')[1])
df['S3'] = df['Angles'].apply(lambda x: x.split(':')[2])
df['XX'] = df['XY'].apply(lambda x: x.split(' ')[0])
df['YY'] = df['XY'].apply(lambda x: x.split(' ')[1])
df.head(125)
```

```python
# From 78th angle, swap S3 and S1. Backup S3 to a temporary column
df['S3_temp'] = df['S3']
df.loc[index:, 'S3'] = df.loc[index:, 'S1']
df.loc[index:, 'S1'] = df.loc[index:, 'S3_temp']
df = df.drop(columns=['S3_temp'])
df.head(125)
```

```python
# Saved to cleaned_data.csv
df.to_csv('cleaned_data_2.csv', index=False)
```
